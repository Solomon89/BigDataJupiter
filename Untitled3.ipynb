```python
import pandas
from sklearn import linear_model
import numpy as np
df = pandas.read_csv("C:\\Users\\Дмитрий\\Downloads\\cars.csv")
df
```

```python
X = df[['Weight', 'Volume']]
y = df['CO2']
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)

predictedCO2 = regr.predict([[2300, 1300]])
print(predictedCO2)
```

```python
print(regr.coef_)
```
