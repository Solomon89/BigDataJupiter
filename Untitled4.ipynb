```python
import pandas
from sklearn import linear_model
import numpy as np
df = pandas.read_csv("C:\\Users\\Дмитрий\\Downloads\\states.csv")
df
```

```python
df.describe()
```

```python
X = df[['metro_res','white','hs_grad','female_house']]
y = df['poverty']
X
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
r_sq = regr.score(X, y)
print('coefficient of determination:', r_sq)
print('intercept:', regr.intercept_)
print('slope:', regr.coef_)
```

```python
nameArray = ['metro_res','white','hs_grad','female_house','poverty'] 
rez = []
i = 1
rez.append( [' element ','metro_res','white','hs_grad','female_house','poverty'] )
for nameX in nameArray:
    rez.append([])
    rez[i].append(nameX) 
    for nameY in nameArray:
        regr = linear_model.LinearRegression()
        regr.fit(df[[nameX]],df[nameY])
        r_sq = regr.score(df[[nameX]],df[nameY])
        #print('coefficient of determination:', r_sq)
        #print('intercept:', regr.intercept_)
        rez[i].append(regr.coef_[0])
    i+=1
np.array(rez)

for item in rez:
    line = ""
    for element in item:
        line = str(element) + "\t "
        print(line)
```
